```python
import igraph as ig
import matplotlib.pyplot as plt
import numpy as np
import cdlib.algorithms






g = ig.Graph.Read_GraphML("marvel-network.graphml")
vertices = g.vs
edges =g.es
label = g.vs[0]["label"]
weight = g.es[0]["weight"]
plot =ig.plot(g)
plot.save("myplot.png")
```

```python
# L’analyse de la distribution des degrés
degrees = g.degree()
plt.hist(degrees, bins=range(max(degrees)+1))
plt.xlabel("Degree")
plt.ylabel("Frequency")
plt.show()
```

```python
#L’analyse des composants connectés
# Get the connected components
components = g.components()
# Get the sizes of the connected components
sizes = components.sizes()
membership = components.membership
print("Sizes of the connected components:", sizes)
```

```python
# obtenir le plus court chemin entre deux sommets
# get the shortest distances from node 1 to all other nodes
distances = g.distances(source=1, target=None, weights=None)
print("distances : ", distances)
# calculate betweenness centrality for all vertices
bc = g.betweenness()
print("        ")
print("Betweenness centrality for all vertices: ", bc)
```

```python
#Le coefficient de clustering et l’analyse de la densité
avg_clustering = g.transitivity_undirected()
density = g.density()
print("Le coefficient de clustering moyen est : ", avg_clustering)
print("La densité du graphe est : ", density)
```

```python
# Calculer la centralité de degré
deg_centrality = g.degree()
print("centralité des noeuds:",deg_centrality)
```

```python
from igraph import Graph
from igraph import plot
from igraph import VertexClustering

# Trouvez le partitionnement avec Louvain
partition = g.community_multilevel()

# Affichez le nombre de communautés trouvées
print("Nombre de communautés détectées : ", len(partition))

# Affichez la partition des nœuds
print("Partition des noeuds : ", partition.membership)

# Affichez le graphe avec les communautés colorées
visual_style = {}
visual_style["vertex_color"] = partition.membership
plot(g, **visual_style)
```

```python
import igraph as ig

# créer un graphe aléatoire
g = ig.Graph.Erdos_Renyi(n=100, m=300)

# appliquer l'algorithme Demon/Angel
coms = g.community_fastgreedy().as_clustering()

# afficher les communautés identifiées
for i, c in enumerate(coms):
    print(f"Communauté {i+1}: {c}")

# visualiser le graphe avec les communautés identifiées
visual_style = {}
visual_style["vertex_color"] = coms.membership
ig.plot(g, **visual_style)
```

```python
# appliquer l'algorithme Propagation des labels
coms = g.community_label_propagation()

# afficher les communautés identifiées
for i, c in enumerate(coms):
    print(f"Communauté {i+1}: {c}")

# visualiser le graphe avec les communautés identifiées
visual_style = {}
visual_style["vertex_color"] = coms.membership
ig.plot(g, **visual_style)
```
